```python
cpu["Core_Speed"] = cpu["Core_Speed"].replace("\n- ", np.nan)
cpu.rename(columns={"Core_Speed": "Core_Speed_(MHz)"}, inplace=True)
cpu["Core_Speed_(MHz)"] = cpu["Core_Speed_(MHz)"].str.replace("MHz", "")
cpu["Core_Speed_(MHz)"] = pd.to_numeric(cpu["Core_Speed_(MHz)"], errors="coerce")  # ép về float
```

```python
cpu["Direct_X"] = cpu["Direct_X"].str.replace("DX", "")
cpu.rename(columns={"Direct_X": "Direct_X_(DX)"}, inplace=True)
cpu["Direct_X_(DX)"] = pd.to_numeric(cpu["Direct_X_(DX)"], errors="coerce")  # type cast to float
```

```python
cpu["L2_Cache"] = cpu["L2_Cache"].str.replace("KB", "")
cpu["L2_Cache"] = pd.to_numeric(cpu["L2_Cache"], errors="coerce")  # type cast to float
cpu.rename(columns={"L2_Cache": "L2_Cache_(KB)"}, inplace=True)
```

```python
cpu["Max_Power"] = cpu["Max_Power"].str.replace("Watts", "")
cpu["Max_Power"] = pd.to_numeric(cpu["Max_Power"], errors="coerce")  # type cast to float
cpu.rename(columns={"Max_Power": "Max_Power_(Watts)"}, inplace=True)
```

```python
cpu["Memory"] = cpu["Memory"].str.replace("MB", "")
cpu["Memory"] = pd.to_numeric(cpu["Memory"], errors="coerce")  # type cast to float
cpu.rename(columns={"Memory": "Memory_(MB)"}, inplace=True)
```

```python
def drop_cols_with_nan(df, threshold=0.5):
    """
    remove columns if ratio NaN/(total rows) > threshold.
    default threshold = 0.5 (50%).
    """
    return df.dropna(axis=1, thresh=len(df) * (1 - threshold))
```

```python
cpu = drop_cols_with_nan(cpu, threshold=0.5)
cpu.shape
```

```python
gpu = drop_cols_with_nan(gpu, threshold=0.5)
gpu.shape
```

```python
# Encode for caluculation purpose. Yes --> 1, No --> 0 for all columns
cpu = cpu.replace({"Yes":1, "No": 0})
gpu = gpu.replace({"Yes": 1, "No": 0})
```
