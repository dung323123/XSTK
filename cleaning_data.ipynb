```python
import pandas as pd
import numpy as np
```

```python
gpu = pd.read_csv("All_GPUs.csv", encoding = "utf-8")
pd.set_option('display.max_columns', None)  # To see all columns of dataframe
```

```python
gpu.describe() # To see basic statistical data of numeric columns
```

```python
gpu.info() # To see features of dataframe
```

```python
gpu["Core_Speed"] = gpu["Core_Speed"].replace("\n- ", np.nan)
gpu.rename(columns={"Core_Speed": "Core_Speed_(MHz)"}, inplace=True)
gpu["Core_Speed_(MHz)"] = gpu["Core_Speed_(MHz)"].str.replace("MHz", "")
gpu["Core_Speed_(MHz)"] = pd.to_numeric(gpu["Core_Speed_(MHz)"], errors="coerce")  # ép về float
```

```python
gpu["Direct_X"] = gpu["Direct_X"].str.replace("DX", "")
gpu.rename(columns={"Direct_X": "Direct_X_(DX)"}, inplace=True)
gpu["Direct_X_(DX)"] = pd.to_numeric(gpu["Direct_X_(DX)"], errors="coerce")  # type cast to float
```

```python
gpu["L2_Cache"] = gpu["L2_Cache"].str.replace("KB", "")
gpu["L2_Cache"] = pd.to_numeric(gpu["L2_Cache"], errors="coerce")  # type cast to float
gpu.rename(columns={"L2_Cache": "L2_Cache_(KB)"}, inplace=True)
```

```python
gpu["Max_Power"] = gpu["Max_Power"].str.replace("Watts", "")
gpu["Max_Power"] = pd.to_numeric(gpu["Max_Power"], errors="coerce")  # type cast to float
gpu.rename(columns={"Max_Power": "Max_Power_(Watts)"}, inplace=True)
```

```python
gpu["Memory"] = gpu["Memory"].str.replace("MB", "")
gpu["Memory"] = pd.to_numeric(gpu["Memory"], errors="coerce")  # type cast to float
gpu.rename(columns={"Memory": "Memory_(MB)"}, inplace=True)
```

```python
gpu["Memory_Bandwidth_(GB/sec)"] = pd.to_numeric(
    gpu["Memory_Bandwidth"].str.replace("GB/sec", "", regex=False).str.strip(),
    errors="coerce"
)
gpu.drop(columns="Memory_Bandwidth", inplace=True)
```

```python
gpu["Memory_Bus_(Bit)"] = pd.to_numeric(
    gpu["Memory_Bus"].str.replace("Bit", "", regex=False).str.strip(),
    errors="coerce"
)
gpu.drop(columns="Memory_Bus", inplace=True)
```

```python
gpu["Memory_Speed_(MHz)"] = pd.to_numeric(
    gpu["Memory_Speed"].str.replace("MHz", "", regex=False).str.strip(),
    errors="coerce"
)
gpu.drop(columns="Memory_Speed", inplace=True)
```

```python
gpu["Open_GL"] = pd.to_numeric(gpu["Open_GL"], errors="coerce")
```

```python
psu_split = gpu["PSU"].str.extract(r"(\d+)\s*Watt.*?(\d+)\s*Amps")

gpu["PSU_Watt_(Watt)"] = pd.to_numeric(psu_split[0], errors="coerce")
gpu["PSU_Amps_(Amps)"] = pd.to_numeric(psu_split[1], errors="coerce")

gpu.drop(columns="PSU", inplace=True)
```

```python
gpu["Pixel_Rate_(GPixel/s)"] = pd.to_numeric(
    gpu["Pixel_Rate"].str.replace("GPixel/s", "", regex=False).str.strip(),
    errors="coerce"
)
gpu.drop(columns="Pixel_Rate", inplace=True)
```

```python
gpu["Process_(nm)"] = pd.to_numeric(
    gpu["Process"].str.replace("nm", "", regex=False).str.strip(),
    errors="coerce"
)
gpu.drop(columns="Process", inplace=True)
```

```python
gpu["ROPs"] = pd.to_numeric(gpu["ROPs"], errors="coerce")
gpu["Shader"] = pd.to_numeric(gpu["Shader"], errors="coerce")
gpu["TMUs"] = pd.to_numeric(gpu["TMUs"], errors="coerce")
```

```python
gpu["Release_Date"] = gpu["Release_Date"].str.strip()  # Bỏ ký tự thừa như "\n"
gpu["Release_Date"] = pd.to_datetime(gpu["Release_Date"], errors="coerce", format="%d-%b-%Y")
```

```python
res_split = gpu["Resolution_WxH"].str.split("x", expand=True)
gpu["Resolution_Width"] = pd.to_numeric(res_split[0], errors="coerce")
gpu["Resolution_Height"] = pd.to_numeric(res_split[1], errors="coerce")
gpu.drop(columns="Resolution_WxH", inplace=True)
gpu.head(5)
```

```python
gpu["Texture_Rate_(GTexel/s)"] = pd.to_numeric(
    gpu["Texture_Rate"].str.replace("GTexel/s", "", regex=False).str.strip(),
    errors="coerce"
)
gpu.drop(columns="Texture_Rate", inplace=True)
```

```python
gpu["Release_Price_(USD)"] = pd.to_numeric(
    gpu["Release_Price"].str.replace("$", "", regex=False).str.strip(),
    errors="coerce"
)
gpu.drop(columns="Release_Price", inplace=True)
```

```python
non_null_values = gpu[gpu["Boost_Clock"].notna()]["Boost_Clock"]
print(non_null_values)
```

```python
gpu["Boost_Clock_(MHz)"] = pd.to_numeric(
    gpu["Boost_Clock"].str.replace("MHz", "", regex=False).str.strip(),
    errors="coerce"
)
gpu.drop(columns="Boost_Clock", inplace=True)
```

```python
def drop_cols_with_nan(df, threshold=0.5):
    """
    remove columns if ratio NaN/(total rows) > threshold.
    default threshold = 0.5 (50%).
    """
    return df.dropna(axis=1, thresh=len(df) * (1 - threshold))
```

```python
gpu = drop_cols_with_nan(gpu, threshold=0.1)
gpu.shape
```

```python
# Encode for caluculation purpose. Yes --> 1, No --> 0 for all columns
gpu = gpu.replace({"Yes": 1, "No": 0})
```
